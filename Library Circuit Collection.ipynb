```python
from qiskit import transpile
```

```python
def show(circuit, **kwargs):
    if hasattr(circuit, '_build'):
        circuit._build()
    kwargs['output'] = 'mpl'
    basis_gates = ['id', 'x', 'y', 'z', 'h', 's', 't', 'sdg', 'tdg', 'rx', 'ry', 'rz',
                   'rxx', 'ryy', 'cx', 'cy', 'cz', 'ch', 'crx', 'cry', 'crz', 'swap',
                   'cswap', 'ccx', 'cu1', 'cu3', 'u1', 'u2', 'u3', 'dcx', 'rzx', 'rzz']
    return transpile(circuit, basis_gates=basis_gates, optimization_level=0).draw(**kwargs)
```

```python
from qiskit.circuit.library import QFT
```

```python
show(QFT(4, insert_barriers=True), filename='qft.pdf')
```

```python
show(QFT(4).inverse(), filename='iqft.pdf')
```

```python
from numpy import random, sign
```

```python
f = [int(sign(r)) for r in random.random(2**3) - 0.5]
g = [int(sign(r)) for r in random.random(2**3) - 0.5]

show(FourierChecking(f, g), filename='fourier_checking.pdf')
```
